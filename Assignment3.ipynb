```python
#in case we want to access the file without unzipping it

import io
from zipfile import ZipFile
import pandas as pd



def read_zip(zip_fn, extract_fn=None):
    zf = ZipFile(zip_fn)
    if extract_fn:
        return zf.read(extract_fn)
    else:
        return {name:zf.read('GLM_data/Table 1.1 Tropical cyclones.xls') for name in zf.namelist()}
#d = ZipFile.extract('GLM_data/Table 1.1 Tropical cyclones.xls')
data = pd.read_excel(io.BytesIO(read_zip(r'C:\Users\tass4\Desktop\statistics for DS\GLM_data.zip', 'GLM_data/Table 1.1 Tropical cyclones.xls')), skiprows = 2)
data
```

```python
#Reading MATLAB files
#imports
import pandas as pd
import numpy as np
from scipy.io import loadmat
from scipy.io import savemat
import matplotlib.pyplot as plt

#load the mat file..
mat = loadmat(r'C:\Users\tass4\Desktop\statistics for DS\test_matlab.mat')
#get the variables
xdata=mat['x']
ydata=mat['y']
zdata=mat['z']

#let y be a column vector 
ydata=np.transpose(ydata)

#plotting

cp=plt.pcolor(xdata,ydata,zdata,cmap='jet', vmin=0,vmax=5000)
plt.colorbar(cp)
plt.title('plotting from a MAT file')
plt.xlabel('x-label')
plt.ylabel('y-label')
```
